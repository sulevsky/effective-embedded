```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

### Total

```python
df = pd.read_csv("experiment_results/output.csv")

```

```python
df.describe()
```

### Experiments

```python
attenuations = [
    "0dB",
    "2.5dB",
    "6dB",
    "12dB",
]
for (exp_id, attenuation) in enumerate(attenuations, 1):
    exp_df = df[df["experiment_id"] == exp_id]
    print(f"Experiment {exp_id}, attenuation: {attenuation}")
    print(f"Number of samples: {len(exp_df)}")
    print(f"Max voltage: {max(exp_df["calibrated"])}")
    plt.plot(exp_df["raw"], exp_df["calibrated"], marker='.')
    plt.xlabel("Raw ADC (bin)")
    plt.ylabel("Calibrated (mV)")
    plt.title(f"Calibrated vs Raw ADC - {exp_id} ({attenuation})")
    plt.grid(True)
    plt.show()
```
